```python
currencies = {'India': 'Rupees', 'Japan':'Yen', 'USA':'Dollar'}
pd.Series(currencies) # series using a dictionary, keys become indexes
```

```python
pd.Series(animals, index=nums) # specifying the indexes and values separately for a series
```

```python
# overriding the keys in the dictionary and only using values for specifieed keys, unspecified keys are assigned NaN
pd.Series(currencies, index=['India', 'USA', 'UK']) 
```

```python
s = pd.Series(currencies)
s
```

```python
s.iloc[2] # indexing using iloc index location
```

```python
s.loc['Japan'] # indexing using loc index name
```

```python
s = pd.Series(np.random.randint(0, 10000, 10000)) # series with 10000 random numbers
s.head()
```

```python
len(s)
```

```python
%%timeit -n 100

tot = 0
for n in s:
    tot += n  # slower way of calculating sum of a series
print(tot)
```

```python
%%timeit -n 100

total = np.sum(s)  # faster way of calculating sum of a series
print(total)
```

```python
s += 2  # simply adding 2 to each element of the series
s.head()
```

```python
sports = pd.Series({'Archery':'Bhutan', 'Sumo':'Japan', 'Taekwondo':'South Korea'})
sports
```

```python
sports.loc['Golf'] = 'Scotland' # using loc to add key value to the series
sports
```

```python
# series having indexes that are not unique
cricket = pd.Series(['India', 'Pakistan', 'England'], index=['Cricket', 'Cricket', 'Cricket'])
cricket                    
```

```python
alls = sports.append(cricket) # appending the cricket values to the sport
```

```python
sports # series is immutable, so doesnt change the original series
```

```python
alls # values get appended in a new series altogether
```
