```python
import pandas as pd
```

```python
odf = pd.read_csv('olympics.csv', index_col = 0, skiprows = 1)
odf.head()
```

```python
# renaming the columns from the olympics dataset using a for loop and ladder if else
for col in odf.columns:
    if col[:2] == '01':
        odf.rename(columns={col:'Gold'+col[4:]}, inplace = True)
    elif col[:2] == '02':
        odf.rename(columns={col:'Silver'+col[4:]}, inplace = True)
    elif col[:2] == '03':
        odf.rename(columns={col:'Bronze'+col[4:]}, inplace = True)
odf.head()
```

```python
odf['Gold'] > 0 # comparing a column with a particular value
```

```python
type(odf['Gold'] > 0) # retruns a series that has boolean values for the comparison
```

```python
# where is used to overlay a boolean mask on the dataframe/series and return the rows of data that satisfy the condition
only_gold = odf.where(odf['Gold'] > 0)
only_gold.head()
```

```python
only_gold['Gold'].count() # even though it has NaN in the display, it only counts the one with valid data
```

```python
only_gold = only_gold.dropna() # used to drop the NaN rows
only_gold.head()
```

```python
# instead of using where, we can pass the boolean condition in the indexing operator, this removes the NaN rows as well
gold_only = odf[odf['Gold'] > 0]
gold_only.head()
```

```python
# countries that have atleast one gold in summer as well as winter olympics
len(odf[(odf['Gold'] > 0) | (odf['Gold.1'] > 0)])
```

```python
# countries that have atleast one gold in winter but no gold in summer
odf[(odf['Gold'] == 0) & (odf['Gold.1'] > 1)]
```

```python
odf.index # gets the index of the dataframe
```

```python
odf['Country'] = odf.index # make the index into a particular column
odf.index = odf['Gold'] # assign the index as another column
odf.head()
```

```python
# extra empty line is printed since the index column now has a name
odf['Gold'] = odf.index # make the index into a particular column
odf = odf.set_index('Country') # assign the index as another column using set_index
odf.head()
```

```python
odf.reset_index() # resets the indexing to a numbered sequence
odf.head()
```

```python
census = pd.read_csv('census.csv')
census.head()
```

```python
# list of columns that need to be focussed on
columns_needed = ['STNAME',
                  'CTYNAME',
                  'BIRTHS2010',
                 'BIRTHS2011',
                  'BIRTHS2012',
                  'BIRTHS2013',
                  'BIRTHS2014',
                  'BIRTHS2015',
                 'POPESTIMATE2010',
                 'POPESTIMATE2011',
                 'POPESTIMATE2012',
                 'POPESTIMATE2013',
                 'POPESTIMATE2014',
                 'POPESTIMATE2015',]
```

```python
# creating a dataframe with the list of columns needed
cdf = census[columns_needed]
cdf.head()
```

```python
# indexing using 2 columns, like composite keys in RDBMS
cdf = cdf.set_index(['STNAME', 'CTYNAME'])
cdf.head()
```
