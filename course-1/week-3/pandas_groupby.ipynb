```python
import pandas as pd
import numpy as np
df = pd.read_csv('census.csv')
df = df.where(df['SUMLEV'] == 50)
```

```python
# to get the average population of all the counties in a particular state
# we loop through for each state and calculate the average
# this takes more time
%%timeit -n 10

for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME'] == state).dropna()['CENSUS2010POP'])
    print('Average population of', state, 'is',avg)
```

```python
# to get the average population of all the counties in a particular state
# we loop through for each state and calculate the average
# this takes less time
# %%timeit -n 10

for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Average population of', group, 'is',avg)
```

```python
# we can create our customized grouping function which will be applied on the index
# the function below, segregates the indices of the dataframe based on the remainder with 3
def fun(item):
    if (item % 3) == 0:
        return 0
    if (item % 3) == 1:
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('# of items in',group,' are:',len(frame))
```

```python
# the aggregate function takes a dictionary of column name as keys and the fucntion to be applied on that column as value
# the below dataframe created, gives the average CENSUS2010POP for each state and\
# the maximum POPESTIMATE2010 for each county in each state
df.groupby('STNAME').agg({'CENSUS2010POP': np.max, 'POPESTIMATE2010': np.max})
```

```python
# using the level in group by when the dataframe has multiple indexes
copy_df = df.copy()
copy_df.set_index(['STNAME', 'CTYNAME'], inplace = True)
for group, frame in copy_df.groupby(level=1):
    print('# of records in', group, 'are:', len(frame))
```

```python
# renaming the column names in the dictionary will apply the function on both columns
df.groupby(level=0)['POPESTIMATE2010', 'CENSUS2010POP'].agg({'SUM': np.sum, 'AVG': np.average}).head()
```

```python
# when we keep the names same in the dictionary then the respective functions are applied to the columns
df.groupby(level=0)['POPESTIMATE2010', 'CENSUS2010POP'].agg({'POPESTIMATE2010': np.sum, 'CENSUS2010POP': np.average}).head()
```
