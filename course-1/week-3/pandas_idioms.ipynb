```python
import pandas as pd
```

```python
df = pd.read_csv('census.csv')
df.head()
```

```python
# method chaining is always much better than index chaining and is considered "pandorable"
# execution time wise, it is not efficient, but it is more readable
(df.where(df['SUMLEV'] == 40)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'REGION': 'Region'})
    .head())
```

```python
# traditional way of performing the same task as above
# execution time wise it is efficient, but not very readable
cdf = df[df['SUMLEV'] == 40]
cdf.set_index(['STNAME', 'CTYNAME'], inplace = True)
cdf.rename(columns={'REGION': 'Region'}, inplace = True)
cdf.head()
```

```python
student_1 = pd.Series({'Name':'Shashwat',
                        'Course':'FAI',
                       'Grade':85})
student_2 = pd.Series({'Name':'Shashwat',
                        'Course':'IR',
                       'Grade':96})
student_3 = pd.Series({'Name':'Aatish',
                        'Course':'CSS',
                       'Grade':92})
students = pd.DataFrame([student_1, student_2, student_3], index = ['Student 1', 'Student 1', 'Student 3'])
students
```

```python
# to drop the students who got below 90 in any subject 
students.drop(students[students['Grade'] < 90].index)
```

```python
import numpy as np
import random as r
```

```python
# this function takes a row and finds the minimum and maximum POPESTIMATES from 2010-2015
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'MIN': np.min(data), 'MAX': np.max(data)})
```

```python
# the apply function applies the min_max function to every row in the dataframe
# axis = 0 is for column wise operation and axis = 1 is for row wise operation
cdf.apply(min_max, axis = 1).head()
```

```python
# using the lambda to print squares of every element of a list
l = []
for i in range(10):
    l.append(r.randint(0,50))
print(l)    
list(map((lambda x: x**2), l))
```

```python
# finding the maximum popestimates from 2010-2015 using lambda
pop_rows = ['POPESTIMATE2010',
             'POPESTIMATE2011',
             'POPESTIMATE2012',
             'POPESTIMATE2013',
             'POPESTIMATE2014',
             'POPESTIMATE2015']
cdf.apply((lambda x: np.max(x[pop_rows])), axis = 1)
```
